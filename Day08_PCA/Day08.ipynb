```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
y = digits.target
```

```python
print(f"Rows : {X.shape[0]}, Columns : {X.shape[1]}")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
df_pca = pd.DataFrame(X_pca, columns = ["PC1", "PC2"])
df_pca["label"] = y
df_pca.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="PC1", y="PC2", hue="label", data=df_pca, palette="tab10", alpha=0.7)
plt.title("MNIST : 64 Dimensions Compressed to 2")
plt.show()
```

```python
variance = np.sum(pca.explained_variance_ratio_) * 100
print(f"Information Retained: {variance:.2f}%")
```
